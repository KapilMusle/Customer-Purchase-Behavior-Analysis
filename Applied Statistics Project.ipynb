##Task 2 - Descriptive Statistics üìä

- **Calculate measures of central tendency (mean, median, mode) and measures of dispersion (variance, standard deviation) for key variables. Identify and handle outliers if necessary.**


```python
df.describe()
```

```python
# Define key variables
key_variables = ['Income', 'Year_Birth', 'Kidhome', 'Teenhome', 'Recency',
                 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts',
                 'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases',
                 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases',
                 'NumWebVisitsMonth', 'Response', 'Complain']

# Calculate measures of central tendency
central_tendency = df[key_variables].agg(['mean', 'median', lambda x: x.mode().iloc[0]])

# Calculate measures of dispersion
variance = df[key_variables].var()
std_deviation = df[key_variables].std()

# Display the results
print("Measures of Central Tendency (Mean, Median, Mode):")
print(central_tendency)
print("\nVariance:")
print(variance)
print("\nStandard Deviation:")
print(std_deviation)
```



Below code will exclude non-numeric columns when calculating descriptive statistics, to prevent 'Graduation' value (or any other non-numeric values) from causing errors.

```python
import pandas as pd

# Exclude non-numeric columns
numeric_columns = df.select_dtypes(include=['number'])

# Calculate descriptive statistics
descriptive_stats = numeric_columns.describe()

# Calculate mode separately for each variable
mode = numeric_columns.mode().iloc[0]

# Calculate variance separately for each variable
variance = numeric_columns.var()

# Calculate standard deviation separately for each variable
std_deviation = numeric_columns.std()

# Display the descriptive statistics
print("Descriptive Statistics:")
print(descriptive_stats)

# Display the mode for each variable
print("\nMode:")
print(mode)

# Display the variance for each variable
print("\nVariance:")
print(variance)

# Display the standard deviation for each variable
print("\nStandard Deviation:")
print(std_deviation)
```

## Task 3 - Probability Distributions üé≤

- **Identify variables that could follow specific probability distributions (e.g., Binomial, Normal). Calculate probabilities and expected values based on these distributions.**



**Deliverables**:

- **Determination of suitable probability distributions for relevant variables and corresponding calculated probabilities and expected values.**:

  

**Binomial Distribution for 'Complain' Variable:**


*  Complain: Complain variable represents whether a customer have complained in the last 2 years (1 if complained, 0 otherwise).

```python
from scipy.stats import binom

# Probability of a customer complaining (complain=1)
p_complain = df['Complain'].mean()

# Expected number of customers complaining
e_complainers = len(df) * p_complain

print(f"Probability of a customer complaining customer : {p_complain:.4f}")
print(f"Expected number of complainers: {e_complainers:.2f}")
```

**Poisson Distribution for 'NumDealsPurchases' Variable:**

*   This variable represents the number of purchases made with a discount, which could potentially follow a Poisson distribution if the number of events (discount purchases) is rare and independent.
*   We fit a Poisson distribution to the 'NumDealsPurchases' data and calculate the probability of a specific number of discount purchases (e.g., 2).



```python
from scipy.stats import poisson

# Fit a Poisson distribution to the NumDealsPurchases data
mu_purchases = df['NumDealsPurchases'].mean()

# Probability of a specific number of discount purchases (e.g., 2)
probability_2_discount_purchases = poisson.pmf(2, mu_purchases)

print(f"Mean of discount purchases: {mu_purchases:.2f}")
print(f"Probability of 2 discount purchases: {probability_2_discount_purchases:.4f}")
```

```python
import seaborn as sns

# Plot relationship between age and spending on wines
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Year_Birth', y='MntWines', data=df, hue='Education', palette='Set2')
plt.title('Relationship between Age and Spending on Wines by Education Level')
plt.xlabel('Year of Birth')
plt.ylabel('Spending on Wines')
plt.legend(title='Education')
plt.grid(True)
plt.show()
```

```python

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans


# Preprocess the data
df = df.dropna(subset=['Year_Birth', 'Income'])

# Ensure the columns are numeric
df.loc[:, 'Year_Birth'] = pd.to_numeric(df['Year_Birth'], errors='coerce')
df.loc[:, 'Income'] = pd.to_numeric(df['Income'], errors='coerce')

# Drop any rows with NaN values in 'Year_Birth' or 'Income' columns
df = df.dropna(subset=['Year_Birth', 'Income'])

# Normalize the data for clustering
scaler = StandardScaler()
df[['Year_Birth', 'Income']] = scaler.fit_transform(df[['Year_Birth', 'Income']])

# Define the number of clusters
num_clusters = 3

# Apply K-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(df[['Year_Birth', 'Income']])

# Segment customers based on age and income
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Year_Birth', y='Income', data=df, hue='Cluster', palette='Set1', style='Cluster', s=100)
plt.title('Customer Segmentation based on Year of Birth and Income')
plt.xlabel('Year of Birth')
plt.ylabel('Income')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()

# Make recommendations
print("Recommendations:")
print("- Target marketing campaigns towards younger customers with higher income, as they represent a lucrative segment for wine purchases.")
print("- Tailor promotions and discounts based on customer segmentation to incentivize spending and increase customer engagement.")
print("- Collect feedback from customers with different education levels to understand their preferences and tailor product offerings accordingly.")
```
